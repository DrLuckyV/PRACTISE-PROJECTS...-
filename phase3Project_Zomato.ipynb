## Data Storage:
This problem statement contains two Datasets- <b>Zomato.csv</b> and <b>country_code.csv.</b>
<b>Country_code.csv</b> contains two variables:
1. Country Code
2. Country Name

The collected Data has been stored in the Comma Separated Value file **Zomato.csv.** Each
restaurant in the Dataset is uniquely identified by its Restaurant Id. Every Restaurant contains the following variables:

1. <b>Restaurant Id</b>: Unique id of every restaurant across various cities of the world
2. <b>Restaurant Name</b>: Name of the restaurant
3. <b>Country Code</b>: Country in which restaurant is located
4. <b>City</b>: City in which restaurant is located
5. <b>Address</b>: Address of the restaurant
6. <b>Locality</b>: Location in the city
7. <b>Locality Verbose</b>: Detailed description of the locality
8. <b>Longitude</b>: Longitude coordinate of the restaurant&#39;s location
9. <b>Latitude</b>: Latitude coordinate of the restaurant&#39;s location
10. <b>Cuisines</b>: Cuisines offered by the restaurant
11. <b>Average Cost for two</b>: Cost for two people in different currencies ��
12. <b>Currency</b>: Currency of the country
13. <b>Has Table booking</b>: yes/no
14. <b>Has Online delivery</b>: yes/ no
15. <b>Is delivering</b>: yes/ no
16. <b>Switch to order menu</b>: yes/no
17. <b>Price range</b>: range of price of food
18. <b>Aggregate Rating</b>: Average rating out of 5
19. <b>Rating color</b>: depending upon the average rating color
20. <b>Rating text</b>: text on the basis of rating of rating
21. <b>Votes</b>: Number of ratings casted by people

<b>Problem statement : In this Dataset predict 2 things –:</b>
    
   1)Average Cost for **two**
   
   2)Price **range**

# <p style="background-color:#FFFF00;font-family:newtimeroman;color:#000000;font-size:100%;text-align:center;border-radius:10px 10px;">Popular cuisines in India</p>

```python
d1 = indian_exposure.copy()
```

```python
d1["Cuisines"]=d1["Cuisines"].str.split(",")
```

```python
d1.head()
```

```python
## Now we will find out list of all the cuisines available in india.
```

```python
indian_cuisines = []
for i in d1["Cuisines"]:
    for j in i:
        if j.strip() not in indian_cuisines:
            indian_cuisines.append(j.strip())
            
```

```python
indian_cuisines
```

```python
len(indian_cuisines)
```

There are 90 different types of cuisines are available in india

# <p style="background-color:#00FFFF;font-family:newtimeroman;color:#0000FF;font-size:100%;text-align:center;border-radius:10px 10px;">Let's determine the popularity of each cuisines</p>

```python
all_cuisines = [cuisine.strip() for i in d1["Cuisines"] for cuisine in i]

# Get the counts of each cuisine
cuisine_counts = pd.Series(all_cuisines).value_counts()

# Plot the cuisine popularity
plt.figure(figsize=(15, 6))
sns.barplot(x=cuisine_counts.index, y=cuisine_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Popularity of Cuisines')
plt.tight_layout()
plt.show()
```

```python
cuisine_counts = cuisine_counts[cuisine_counts.values>30]
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x=cuisine_counts.index, y=cuisine_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Popularity of Cuisines in india')
plt.tight_layout()
plt.show()
```

From the above chart we can conclude that the most popular cuisine in india is North indian and followed by chinese.

# <p style="background-color:#00FF00;font-family:newtimeroman;color:#0000FF;font-size:100%;text-align:center;border-radius:10px 10px;">Let's figure out the most popular cuisine of the world</p>

```python
d2 = final_zomato.copy()
```

```python
d2["Cuisines"]=d2["Cuisines"].str.split(",")
all_cuisines_around_world = [cuisine.strip() for i in d2["Cuisines"] for cuisine in i]

# Get the counts of each cuisine
cuisine_counts_world = pd.Series(all_cuisines_around_world).value_counts()
cuisine_counts_world = cuisine_counts_world[cuisine_counts_world.values>30]

# Plot the cuisine popularity
plt.figure(figsize=(15, 8))
sns.barplot(x=cuisine_counts_world.index, y=cuisine_counts_world.values)
plt.xticks(rotation=90)
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Popularity of Cuisines in world')
plt.tight_layout()
plt.show()
```

In the world cuisine italian comes on number 4 spot.
